```python
# Histograms
df['Age'].hist(bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python
df['Billing Amount'].hist(bins=10)
plt.title('Billing Amount Distribution')
plt.xlabel('Billing Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Medical Condition'].value_counts().nlargest(10).plot(kind='bar')
plt.title('Top 10 Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot for categorical variable
df['Blood Type'].value_counts().plot(kind='bar')
plt.title('Blood Type Distribution')
plt.xlabel('Blood Type')
plt.ylabel('Count')
plt.show()
```

```python

df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

```

```python
# Box plot for billing amount by gender
sns.boxplot(x='Gender', y='Billing Amount', data=df)
plt.title('Billing Amount by Gender')
plt.show()

```

```python
plt.figure(figsize=(12, 8))
df['Medical Condition'].value_counts().head(10).plot(kind='bar', color='blue')
plt.title('Top  Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Billing Amount'], alpha=0.5, color='purple')
plt.title('Age vs. Billing Amount')
plt.xlabel('Age')
plt.ylabel('Billing Amount')
plt.grid(True)
plt.show()
```
